# 僅計算專利文件相關keywords及數量

```python
# 僅計算專利文件相關keywords及數量
import requests
import spacy
from collections import Counter
from string import punctuation
import xml.etree.ElementTree as ET

# Download the file
url = "https://raw.githubusercontent.com/NievesLiang/patent_documents/ce8773c8457162279f72b8c4f27cae95b0dc7155/US08438662-20130514.txt"
filename = "US08438662-20130514.txt"

response = requests.get(url)
with open(filename, 'wb') as file:
    file.write(response.content)

# Parse XML and extract text content
tree = ET.parse(filename)
root = tree.getroot()
text = ""
for element in root.iter():
    if element.text:
        text += element.text.strip() + " "

# Process extracted text with spaCy
nlp = spacy.load("en_core_web_sm")

def get_hotwords(text):
    result = []
    pos_tags = ['PROPN', 'NOUN', 'ADJ', 'VERB']  # Updated the POS tags
    doc = nlp(text.lower())
    for token in doc:
        if token.text in nlp.Defaults.stop_words or token.text in punctuation:
            continue
        if token.pos_ in pos_tags:
            result.append(token.text)
    return result

# Define a list of patent-specific terms
patent_terms = ['invention', 'patent', 'claim', 'application', 'novel', 'innovative', 'technology', 'inventor']

output = get_hotwords(text)  # Removed set() since Counter already handles duplicates
most_common_list = Counter(output).most_common(10)
# for item in most_common_list:
    # print(item[0])

# Write extracted text to a readability text file
readability_filename = filename  # Provide a specific name for the readability file

with open(readability_filename, 'w', encoding='utf-8') as file:
    file.write(text)

most_common_dict = dict(most_common_list)
for word, count in most_common_dict.items():
    print(f"{word}: {count}")
```

# 4. Find out two most similar patents. 可使用LLMs( 如Bert, GPT 等transformer models)

# 余弦相似度模型比較(效能較好)

```python
import urllib.request
import re
from transformers import AutoModel, AutoTokenizer
import torch
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

model_name = "bert-base-uncased"
model = AutoModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def compare_txt_similarity(txt_string1, txt_string2):
    # Tokenize the input texts
    inputs1 = tokenizer(txt_string1, padding=True, truncation=True, return_tensors="pt")
    inputs2 = tokenizer(txt_string2, padding=True, truncation=True, return_tensors="pt")

    # Encode the input texts
    with torch.no_grad():
        input_ids1 = inputs1["input_ids"].to(model.device)
        attention_mask1 = inputs1["attention_mask"].to(model.device)
        outputs1 = model(input_ids1, attention_mask=attention_mask1)
        embeddings1 = get_last_hidden_state(outputs1)

        input_ids2 = inputs2["input_ids"].to(model.device)
        attention_mask2 = inputs2["attention_mask"].to(model.device)
        outputs2 = model(input_ids2, attention_mask=attention_mask2)
        embeddings2 = get_last_hidden_state(outputs2)

    # Reduce the dimension of embeddings from 3D to 2D
    embeddings1 = torch.flatten(embeddings1, start_dim=1)
    embeddings2 = torch.flatten(embeddings2, start_dim=1)

    # Convert embeddings to numpy arrays
    embeddings1 = embeddings1.cpu().numpy()
    embeddings2 = embeddings2.cpu().numpy()

    # Compute the cosine similarity between the embeddings
    similarity = cosine_similarity(embeddings1, embeddings2)[0][0]
    return similarity

def get_last_hidden_state(outputs):
    if isinstance(outputs, tuple):  # For models with optional second output
        return outputs[0]
    else:
        return outputs.last_hidden_state

def xml_to_plain_text(xml_string):
    # Remove XML tags using regular expressions
    plain_text = re.sub('<[^<]+?>', '', xml_string)
    return plain_text

def process_txt_file(url):
    # Download the file
    filename = url.split('/')[-1]
    urllib.request.urlretrieve(url, filename)

    # Read the downloaded file and process its contents
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
        processed_text = xml_to_plain_text(text)
        return processed_text

def save_text_to_file(text, filename):
    with open(filename, 'a', encoding='utf-8') as f:
        f.write(text + "\n")

url = "https://raw.githubusercontent.com/NievesLiang/patent_documents/main/US084386{}-20130514.txt"
github_txt = ["US084386{}-20130514.txt".format(i) for i in range(62, 72)]

# Download and process TXT files
txt_files = []
for txt_url in [url.format(i) for i in range(62, 72)]:
    processed_text = process_txt_file(txt_url)
    txt_files.append(processed_text)

# Compare similarity between each pair of TXT files
max_similarity = 0.0
most_similar_files = None

for i in range(len(txt_files)):
    for j in range(i+1, len(txt_files)):
        similarity_score = compare_txt_similarity(txt_files[i], txt_files[j])

        if similarity_score > max_similarity:
            max_similarity = similarity_score
            most_similar_files = (github_txt[i], github_txt[j])

# Save the filenames and similarity score to a file
output_filename = "1234567890.txt"
save_text_to_file("Most similar files: {}".format(most_similar_files), output_filename)
save_text_to_file("Similarity score: {}".format(max_similarity), output_filename)

print("Most similar files:", most_similar_files)
print("Similarity score:", max_similarity)
print("Filenames and similarity score saved to:", output_filename)
```
